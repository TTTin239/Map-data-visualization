#dùng geopandas để đọc shapefile trong /Data/GIS/Population/population_HCMC/population_shapefile/Population_District_Level.shp

```python
import geopandas as gpd
import numpy as np
import folium
import matplotlib.pyplot as plt
```

```python
pop = gpd.read_file("/content/CSL_HCMC/Data/GIS/Population/population_HCMC/population_shapefile/Population_District_Level.shp")
pop
```

#hãy thực hiện vẽ ranh giới các quận lên bản đồ dựa theo hướng dẫn sau:
https://geopandas.readthedocs.io/en/latest/gallery/polygon_plotting_with_folium.html

```python
pop.head()
```

```python
pop.plot(figsize=(6, 6))
plt.show()
```

```python
pop = pop.to_crs(epsg=4326)
print(pop.crs)
pop.head()
```

```python
m = folium.Map(location=[10.6, 107.0], zoom_start=10, tiles='CartoDB positron')
m
```

```python
for _, r in pop.iterrows():
    # Without simplifying the representation of each borough,
    # the map might not be displayed
    sim_geo = gpd.GeoSeries(r['geometry']).simplify(tolerance=0.001)
    geo_j = sim_geo.to_json()
    geo_j = folium.GeoJson(data=geo_j,
                           style_function=lambda x: {'fillColor': 'orange'})
    folium.Popup(r['Dist_Name']).add_to(geo_j)
    geo_j.add_to(m)
m
```

```python
# Project to NAD83 projected crs
pop = pop.to_crs(epsg=2263)

# Access the centroid attribute of each polygon
pop['centroid'] = pop.centroid
```

```python
pop = pop.to_crs(epsg=4326)

# Centroid column
pop['centroid'] = pop['centroid'].to_crs(epsg=4326)

pop.head()
```

```python
for _, r in pop.iterrows():
    lat = r['centroid'].y
    lon = r['centroid'].x
    folium.Marker(location=[lat, lon],
                  popup='length: {} <br> area: {}'.format(r['Shape_Leng'], r['Shape_Area'])).add_to(m)

m
```
